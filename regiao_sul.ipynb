## C√≥digo para agrupamento de 5 em 5 anos

```python
df_table = pd.DataFrame(columns=['Per√≠odo','M√™s/Ano de maior valor', 'Valor', 'M√™s/Ano de menor valor', 'Valor'])
intervalo = 5
```

```python
def preenche_tabela_agrupamento(df_group,intervalo=1):
    maior_valor = ''
    ano_maior = ''
    mes_maior = ''
    menor_valor = ''
    ano_menor = ''
    mes_menor = ''
    ano = ''    
    
    df_group.reset_index(inplace=True,drop=True)
    periodo = '['+str(df_group.iloc[0:1,-1:].values[0][0])+'-'+str(df_group.iloc[0:1,-1:].values[0][0] + intervalo)+']'
    
    for i, row in df_group.iterrows():     
        for j, column in row.iteritems():
            if row.index[-1] == 'Ano':
                ano = str(row.values[-1])
            if (i == 0) & (j == 'Janeiro'):
                maior_valor = column
                menor_valor = column 
                index = f''
            else:            
                if j != 'Ano':
                    if column < menor_valor:
                        menor_valor = column
                        mes_menor = j 
                        ano_menor = ano
                    if column > maior_valor:
                        maior_valor = column
                        mes_maior = j
                        ano_maior = ano
    linhaGroup = [periodo, mes_maior+'/'+ano_maior,maior_valor,mes_menor+'/'+ano_menor,menor_valor]
    return linhaGroup
```

```python
#2006-2010
linha_inicial = 0
df_group = df.drop(columns=['Total']).iloc[linha_inicial:linha_inicial+intervalo+1,:]
df_table.loc[len(df_table)] = preenche_tabela_agrupamento(df_group,intervalo)

#2006-2010
linha_inicial = 6
df_group = df.drop(columns=['Total']).iloc[linha_inicial:linha_inicial+intervalo,:]
df_table.loc[len(df_table)] = preenche_tabela_agrupamento(df_group,intervalo-1)

#2011-2015
linha_inicial = 11
df_group = df.drop(columns=['Total']).iloc[linha_inicial:linha_inicial+intervalo,:]
df_table.loc[len(df_table)] = preenche_tabela_agrupamento(df_group,intervalo-1)

#2016-2020
linha_inicial = 16
df_group = df.drop(columns=['Total']).iloc[linha_inicial:linha_inicial+intervalo,:]
df_table.loc[len(df_table)] = preenche_tabela_agrupamento(df_group,intervalo-1)

df_table
```
