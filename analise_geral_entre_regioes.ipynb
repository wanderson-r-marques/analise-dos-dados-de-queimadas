```python
# Definindo novo dataframe 'df_comparativo_anual'
colunas_comparativo = ['Região']
i = 0

for ano in range(2000,2021,1):
    colunas_comparativo.append(ano)
    if ano < 2020:
        i = ano%1000
        colunas_comparativo.append('%' + '('+ str(i) + '-' + str(i+1) + ')')

df_comparativo_anual = pd.DataFrame(np.zeros((5,42),np.int32),columns=colunas_comparativo)
```

```python
# Definição dos índices (linhas)
regioes = 'Centro-Oeste Nordeste Norte Sudeste Sul'.split()

df_comparativo_anual['Região'] = regioes
df_comparativo_anual.set_index('Região', inplace=True)
```

```python
df_comparativo_anual.head()
```

```python
# Atualização dos valores ano a ano para cada região
for ano in range(2000,2021,1):
    df_comparativo_anual.loc['Centro-Oeste', ano] = df_centro_oeste.loc[ano, 'Total']
    df_comparativo_anual.loc['Nordeste', ano] = df_nordeste.loc[ano, 'Total']
    df_comparativo_anual.loc['Norte', ano] = df_norte.loc[ano, 'Total']
    df_comparativo_anual.loc['Sudeste', ano] = df_sudeste.loc[ano, 'Total']
    df_comparativo_anual.loc['Sul', ano] = df_sul.loc[ano, 'Total']
```

```python
df_comparativo_anual.columns
```

```python
len(df_comparativo_anual.columns)
```

```python
# Cálculo da diferença percentual
ano = 2000

for indice in range(1,40,2):
    coluna = df_comparativo_anual.columns[indice]
    df_comparativo_anual[coluna] = round(((df_comparativo_anual[ano + 1] - df_comparativo_anual[ano])/(df_comparativo_anual[ano]))*100, 1)
    ano += 1
```

```python
df_comparativo_anual.head()
```

```python
# Dataframe contendo apenas o total de focos por ano para cada região
df_total_focos = df_comparativo_anual.iloc[:, 0::2]
```

```python
df_total_focos.head()
```

```python
# Ano com maior concentração de queimadas para cada região 
for coluna in df_total_focos.T.columns:
    print(df_total_focos.T[df_total_focos.T[coluna] == df_total_focos.T[coluna].max()][coluna])
```

```python
# Ano com menor concentração de queimadas para cada região 
for coluna in df_total_focos.T.columns:
    print(df_total_focos.T[df_total_focos.T[coluna] == df_total_focos.T[coluna].min()][coluna])
```

**Resumo dos anos com maior e menor número de focos de incêndio por região**

```python
# Dataframe com as estatísticas descritivas para o total de focos dentro do período analisado para cada região

df_estatisticas_focos= pd.DataFrame(df_total_focos.mean(axis=1).apply(lambda x: int(x)), columns=['Média'])

df_estatisticas_focos['DP'] = df_total_focos.std(axis=1).apply(lambda x: int(x))
df_estatisticas_focos['Mediana'] = df_total_focos.median(axis=1).apply(lambda x: int(x))
df_estatisticas_focos['Máximo'] = df_total_focos.max(axis=1).apply(lambda x: int(x))
df_estatisticas_focos['Mínimo'] = df_total_focos.min(axis=1).apply(lambda x: int(x))
df_estatisticas_focos['Total'] = df_total_focos.sum(axis=1).apply(lambda x: int(x))
```

```python
# Adiciona variação percentual por região em relação ao total de focos
df_estatisticas_focos['Total %'] = round(100 * (df_estatisticas_focos['Total']/df_estatisticas_focos['Total'].sum(axis=0)), 2)
```

**Análise:** 
* Dentro do período avaliado (anos 2000 a 2020), a região que concentrou a maior parte dos focos de incêndio foi a região Norte, respondendo por aproximadamente 41% do total do número de focos de incêndios no Brasil; 
* As regiões com menores concentrações de focos de incêndio são as regiões Sul (4.0%) e Sudeste (6.7%), respectivamente;
* Analisando-se os desvios-padrões, observa-se que os valores alcançados são de no mínimo 35% do valor médio (região Centro-Oeste) e no máximo 56% do valor médio (região Sul). Tais resultados demonstram ampla variabilidade anual das queimadas por região. Destaca-se a região Sul, na qual o desvio-padrão é maior que a metade da média (ou a metade da mediana) mesmo sendo esta a região com menor número de focos para o período avaliado;
* As regiões Centro-Oeste e Nordeste apresentaram resultados totais muito próximos para a quantidade de focos. Cada uma dessas regiões respondeu por aproximadamente 24% do total de queimadas para todo o território brasileiro. Excluindo-se a região Norte, essas duas regiões concentram a maior quantidade de focos e totalizam, aproximadamente, 50% dos focos de incêndios no Brasil;
