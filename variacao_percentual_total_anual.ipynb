```python
import pandas as pd
import numpy as np
```

```python
df_centro_oeste = pd.read_csv("historico_regiao_centro-oeste.csv")
df_nordeste = pd.read_csv("historico_regiao_nordeste.csv")
df_norte = pd.read_csv("historico_regiao_norte.csv")
df_sudeste = pd.read_csv("historico_regiao_sudeste.csv")
df_sul = pd.read_csv("historico_regiao_sul.csv")
```

```python
df_centro_oeste.head()
```

```python
# Remomea coluna referente aos anos
df_centro_oeste.rename(columns={'Unnamed: 0':'Ano'}, inplace=True)
df_nordeste.rename(columns={'Unnamed: 0':'Ano'}, inplace=True)
df_norte.rename(columns={'Unnamed: 0':'Ano'}, inplace=True)
df_sudeste.rename(columns={'Unnamed: 0':'Ano'}, inplace=True)
df_sul.rename(columns={'Unnamed: 0':'Ano'}, inplace=True)
```

```python
df_centro_oeste.shape
#df_nordeste.shape
```

```python
# Remove linhas referentes aos anos de 1998 e 1999 e as linhas 'máximo', 'média' e 'mínimo'
df_centro_oeste.drop([0,1,26,25,24], axis=0, inplace=True)
df_nordeste.drop([0,1,26,25,24], axis=0, inplace=True)
df_norte.drop([0,1,26,25,24], axis=0, inplace=True)
df_sudeste.drop([0,1,26,25,24], axis=0, inplace=True)
df_sul.drop([0,1,26,25,24], axis=0, inplace=True)

# Reinicia os índices no próprio dataframe e não adiciona os índices antigos em uma nova coluna
df_centro_oeste.reset_index(inplace=True, drop=True)
df_nordeste.reset_index(inplace=True, drop=True)
df_norte.reset_index(inplace=True, drop=True)
df_sudeste.reset_index(inplace=True, drop=True)
df_sul.reset_index(inplace=True, drop=True)
```

```python
# Substitui todos os caracteres '-' no dataframe por 0
df_centro_oeste.replace('-', 0, inplace=True)
df_nordeste.replace('-', 0, inplace=True)
df_norte.replace('-', 0, inplace=True)
df_sudeste.replace('-', 0, inplace=True)
df_sul.replace('-', 0, inplace=True)
```

```python
# Verificação dos tipos de dados
df_centro_oeste.dtypes
```

```python
df_nordeste.dtypes
```

```python
# Ajuste dos tipos de dados

df_centro_oeste['Ano'] = df_centro_oeste['Ano'].astype('int16')
df_nordeste['Ano'] = df_nordeste['Ano'].astype('int16')
df_norte['Ano'] = df_norte['Ano'].astype('int16')
df_sudeste['Ano'] = df_sudeste['Ano'].astype('int16')
df_sul['Ano'] = df_sul['Ano'].astype('int16')

# Demais colunas...
colunas = df_centro_oeste.columns[1:]

for coluna in colunas:
    df_centro_oeste[coluna] = df_centro_oeste[coluna].astype('int32')
    df_nordeste[coluna] = df_nordeste[coluna].astype('int32')
    df_norte[coluna] = df_norte[coluna].astype('int32')
    df_sudeste[coluna] = df_sudeste[coluna].astype('int32')
    df_sul[coluna] = df_sul[coluna].astype('int32')
```

```python
# Definindo novo dataframe 'df_comparativo_anual'
colunas_comparativo = ['Região']

for ano in range(2000,2022,1):
    colunas_comparativo.append(ano)
    if ano < 2021:
        colunas_comparativo.append('%' + '('+ str(ano%1000) + ')')

df_comparativo_anual = pd.DataFrame(np.zeros((5,44),np.int32),columns=colunas_comparativo)
```

```python
# Definição dos índices (linhas)
regioes = 'Centro-Oeste Nordeste Norte Sudeste Sul'.split()

df_comparativo_anual['Região'] = regioes
df_comparativo_anual.set_index('Região', inplace=True)
```

```python
df_comparativo_anual.head()
```

```python
# Atualização dos valores ano a ano para cada região
for ano in range(2000,2022,1):
    df_comparativo_anual.loc['Centro-Oeste', ano] = df_centro_oeste.loc[ano, 'Total']
    df_comparativo_anual.loc['Nordeste', ano] = df_nordeste.loc[ano, 'Total']
    df_comparativo_anual.loc['Norte', ano] = df_norte.loc[ano, 'Total']
    df_comparativo_anual.loc['Sudeste', ano] = df_sudeste.loc[ano, 'Total']
    df_comparativo_anual.loc['Sul', ano] = df_sul.loc[ano, 'Total']
```

```python
df_comparativo_anual.columns
```

```python
# Cálculo da diferença percentual
ano = 2000

for indice in range(1,42,2):
    coluna = df_comparativo_anual.columns[indice]
    df_comparativo_anual[coluna] = round(((df_comparativo_anual[ano + 1] - df_comparativo_anual[ano])/(df_comparativo_anual[ano]))*100, 1)
    ano += 1
```
